#### Ниже предоставлен код, который демонстрирует работу программы.
#### В коде присутствует подробное руководство по правильной эксплуатации программы для получения результата. Подсказки будут выводиться на экран при помощи функции print().

```python
from pymystem3 import Mystem  # Импорт библиотеки pymystem3 (модуль Mystem)
import matplotlib.pyplot as plt  # Импорт библиотеки matplotlib.pyplot

m = Mystem()  # Создание экземпляра класса Mystem


def correction(word) -> bool:  # Функция обработки ввода пользователя
    if word == 'да':
        return True
    elif word == 'нет':
        return False
    
def input_word(input_str):     # Функция обработки слов, вводимых с клавиатуры
    global dictionary          # Изменяется не локальная, а глобальная переменная - это надо указать
    if '- это' in input_str and len(input_str) >= 6: # Проверка на правильность ввода ключ.слова и значения
        intermediate_array = input_str.split('- это')  # Разделить строку по дефису и слову 'это'
        dictionary[intermediate_array[0].replace(' ', '')] = ['- это ' + intermediate_array[1],
                                                                      0]  # Добавление ключей и значений в словарь
    else:
        print('Осуществлен неправильный ввод!')

print("Добро пожаловать!\nВыберите одно из дейтсвий:\na. Использовать готовый словарь\nb. Добавить значения в словарь\nc. Использовать только новые слова")
answer = input("Ваш выбор: ")
while answer not in "abc":
    print("Введите букву a, b или c в зависимости от вашего выбора")
    answer = input("Ваш выбор: ")
# Словарик по-умолчанию
dictionary = {"веер": ["- это женственность, хрупкость, таинственность, кокетливость", 0],
                      "забор": ["- это несвобода", 0],
                      "шкаф": ["- это преимущественно - символ России('Вишневый сад')", 0],
                      "шинель": ["- это приспособленчество", 0],
                      "вицмундир": ["- это чинопочитание", 0],
                      "толстый": ["- это скупость, глупость, высокое положение общества", 0],
                      "худощавый": ["- это лицемерие, чинопочитание, пресмыкательство", 0]}
match answer: # Работает по-аналогии с switch-case из языков более низкого уровня.
    case "a": # Если пользователь ввел единичку
        print("Наш словарик остался без изменений")
    case "b": # Если пользователь ввел двоечку
        print(
            "Введите слова и значения в правильном формате!\nПравильный формат: 'Дом- это место жилья'; 'Программирование- это весело'.\nДля прекращения добавления слов в словарь введите пустоту.")
        # Здесь происходит ввод слов с клавиатуры.
        input_str = input("Введите слово и его значение: ")
        while input_str != '':  # Пока пользователь не введет пустую строку
            input_word(input_str)
            input_str = input("Введите слово и его значение: ")  # Повторный запрос ввода значения

    case "c": # Если пользователь ввел троечку
        print(
            "Введите слова и значения в правильном формате!\nПравильный формат: 'Дом- это место жилья'; 'Программирование- это весело'.\nДля прекращения создания словаря введите пустоту.")
        # Здесь происходит ввод слов с клавиатуры.
        dictionary = dict()  # Пустой словарик
        input_str = input("Введите слово и его значение: ")
        while input_str != '':  # Пока пользователь не введет пустую строку
            input_word(input_str)
            input_str = input("Введите слово и его значение: ")  # Повторный запрос ввода значения
print(f"Ваш итоговый вариант словарика выглядит так- {dictionary}")
while True:  # Для конструкции try/except
    file_name = input("Введите название вашего файла: ")
    try:  # Если никаких проблем с открытием нет
        with open(file_name, encoding='utf-8') as fin:  # Открытие файла в кодировке utf-8
            array_line = fin.readlines()  # Запись в переменную все строки файла
        break  # Если успешно прочитано, выходим из цикла
    except FileNotFoundError:  # Ошибка №1 - такого файла нет
        print("Такого файла нет в директории! Проверьте правильность и повторите попытку!")
    except UnicodeDecodeError:  # Ошибка №2 - файл имеет другую кодировку
        print("Файл не может быть прочитан с использованием utf-8. Попробуйте другой файл или кодировку.")
string = ' '.join(array_line).replace("!", '.')  # Обьединить все строки в одну
string = string.replace("?", '.')  # Замена '?' => '.'
string = string.replace("...", '.')  # Замена '...' => '.'
array_sentences = string.split('.')  # Запись в массив предложения по точкам
# Флаг, который нужен для вкл/откл просмотра предложений вручную
flag = True if input(
    'Если вы хотите проверить все предложения вручную, введите "да", иначе любой символ: ') == 'да' else False
for index_sentences in range(len(array_sentences)):  # Пробегаем по каждому предложению
    sentence = array_sentences[index_sentences].split(' ')  # Записываем в переменную предложение ввиде слов
    for index_word in range(len(sentence)):  # Пробегаем по каждому слову в предложении
        lemmas_word = (' '.join(m.lemmatize(sentence[index_word]))).lower()[:-2]  # Лемматизируем слово
        if lemmas_word in dictionary.keys():  # Смотрим, есть ли лемматизированное слово в словаре
            if flag:  # Если включен просмотр всех предложений

                print("Возникли вопросы со словом! Необходимо уточнить, совпадает ли значение слова со значением из "
                      "словаря!")

                print(*sentence)  # Вывод предложения, которое надо проверить
                answer = input("Подходит ли это слово по значению? Ваш выбор: ").lower()
                while answer != 'да' and answer != 'нет':  # Повторять ввод, пока пользователь не введет 'да' или 'нет'
                    answer = input('Непонятный ввод! Введите "да" или "нет": ').lower()
                if correction(
                        answer):  # Вызов функции correction для определения введенного слова ('да' или 'нет'). Обработка слова 'да'
                    print("Слово подсчитано")
                    dictionary[lemmas_word][1] += 1  # Подсчет слова
                else:
                    print("Слово не подсчитано")
            else:  # Если выключен просмотр всех предложений => слово автоматически подходит
                dictionary[lemmas_word][1] += 1  # Подсчет слова
file_name_output = open(input("Введите название нового файла, куда надо сохранить результат: "),
                        'w+')  # Ввод названия файла для сохранения результата работы программы
final_string = ''  # Переменная для 'накопления' ответа
for key in dictionary:  # Пробегаем по каждому ключу в словаре
    final_string += key + dictionary[key][0] + f'\nКол-во слов "{key}" в файле: ' + str(
        dictionary[key][1]) + '\n'  # 'Накопление' ответа для записи в файл
file_name_output.write(final_string)  # Запись в файлик полученного результата
file_name_output.close()  # Закрыть файлик
categories = [key for key in dictionary]  # Данные для графика (ключи - в нашем случае определения)
values = [dictionary[key][1] for key in
          dictionary]  # Данные для графика (значения - в нашем случае кол-во слов в произведении)
plt.bar(categories, values, color='red')  # Создаем столбчатый график
# Добавляем заголовок и метки осей
plt.title('Столбчатый график')
plt.xlabel('Слова')
plt.ylabel('Количество')
plt.show()  # Отображаем график
print('Спасибо за работу! До свидания!')
```
